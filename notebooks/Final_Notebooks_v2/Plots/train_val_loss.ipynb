```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Prettier plotting with seaborn
sns.set(font_scale=1.5)
# Ticks instead of whitegrid in order to demonstrate changes to plot ticks better
sns.set_style("ticks")
```

```python
df_val = pd.read_csv("lstm_val_loss.csv")
df_train = pd.read_csv("lstm_total_loss.csv")
df_val = df_val["Value"]
df_train = df_train["Value"]
```

```python
df_val
```

```python
fig, ax = plt.subplots(figsize=(10, 8), dpi=80)


# for key, value in matlab_cols.items():
plt.plot(df_train,label="Training Loss")
plt.plot(df_val,label="Validation Loss")
ax.set_xlabel("Epoch Number", fontsize=14)
ax.set_ylabel("Loss", fontsize=14)

plt.legend()
plt.savefig("train_val_loss.pdf")

# df_train.plot()
# df_val.plot()
```
